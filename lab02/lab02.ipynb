## 2.3 Training and testing sets

```python
X = mnist["data"]
y = mnist["target"].astype(np.uint8)

y.sort_values(inplace=True)
X.reindex(y.index)
```

```python
X_train, X_test = X[:56000], X[56000:]
y_train, y_test = y[:56000], y[56000:]
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

print("y_train:", np.unique(y_train))
print("y_test:", np.unique(y_test))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # shuffle - default true

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

print("y_train:", np.unique(y_train))
print("y_test:", np.unique(y_test))
```

```python
# reload data
X, y = mnist["data"], mnist["target"].astype(np.uint8)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 2.4 SGD Classifier to detect the digit 0.

```python
y_train_0 = (y_train == 0)
y_test_0 = (y_test == 0)

sgd_clf_0 = SGDClassifier(n_jobs=-1) # using all processors
sgd_clf_0.fit(X_train, y_train_0)
```

```python
y_train_pred = sgd_clf_0.predict(X_train)
y_test_pred = sgd_clf_0.predict(X_test)

acc_train_0 = sum(y_train_pred == y_train_0) / len(y_train_0)
acc_test_0 = sum(y_test_pred == y_test_0) / len(y_test_0)

print("Train accuracy:", acc_train_0)
print("Test accuracy:", acc_test_0)

with open("sgd_acc.pkl", 'wb') as f:
    pickle.dump([acc_train_0, acc_test_0], f)
```

```python
from sklearn.model_selection import cross_val_score
score_0 = cross_val_score(sgd_clf_0, X_train, y_train_0, cv=3, scoring="accuracy")

print("Cross validation score:", score_0)

with open("sgd_cva.pkl", 'wb') as f:
    pickle.dump(score_0, f)
```

## 2.5 SGD Classifier for classifying all digits.

```python
from sklearn.metrics import confusion_matrix
sgd_clf = SGDClassifier(n_jobs=-1)
sgd_clf.fit(X_train, y_train)

train_accuracy = sgd_clf.score(X_train, y_train)
test_accuracy = sgd_clf.score(X_test, y_test)

print("Train accuracy:", train_accuracy)
print("Test accuracy:", test_accuracy)

y_train_pred = sgd_clf.predict(X_train)
conf_matrix = confusion_matrix(y_train, y_train_pred)
print(conf_matrix)
```

```python
y_test_pred = sgd_clf.predict(X_test)
cmx = confusion_matrix(y_test, y_test_pred)

with open("sgd_cmx.pkl", 'wb') as f:
    pickle.dump(cmx, f)

print(cmx)
```
